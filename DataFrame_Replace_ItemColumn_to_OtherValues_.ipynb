```python
import pandas as pd
import random

```

```python
# Exemplo: criando 10 registros com valores alternados de 'Desistente' e 'Graduado'
novos_registros = pd.DataFrame({
    'Target': random.choices(['Desistente', 'Graduado'], k=10)
})
```

```python
novos_registros
```

```python
novos_registros['Target'] = novos_registros['Target'].replace({'Desistente': 0, 'Graduado': 1}) #binário
```
